```python
#процент владеющих python
percentage_py = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum()/ x.count()).sort_values(ascending=False)
percentage_py
```

```python
# создание таблицы с вычисленными значениями
py_df = pd.DataFrame({'NumResp': country_grp['LanguageHaveWorkedWith'].count(),
                     'NumKnowsPy': country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum()),
                     'Percentage': percentage_py})

py_df.sort_values(by = 'Percentage', ascending = False, inplace = True)

#ограничение на репрезентативность
py_df.loc[py_df['NumResp'] >= 100]
```
